### Question 5
Replace the missing values in the **`country`** column using the **mode (most common value)**.

<span style="color:red">(2 marks)</span>


```python
df["country"] = df["country"].fillna(df["country"].mode()[0])
print(df.isnull().sum())
```

### Question 10
Create a **bar chart** showing how many times each country appears in the dataset.  
The chart title must be:

**"Country popularity on Netflix"**

<span style="color:red">(2 marks)</span>

```python
# Plot number of passengers that survived
fig1, ax1 = plt.subplots(figsize=(8, 5))
country_counts = df['country'].value_counts(sort=True).head(10)
country_counts.plot(kind='bar', color=['red', 'yellow'], ax=ax1)
ax1.set_title('Country popularity on Netflix(Top 10)', fontsize=16)
ax1.set_xlabel('Countries', fontsize=12)
ax1.set_ylabel('Count', fontsize=12)
```
